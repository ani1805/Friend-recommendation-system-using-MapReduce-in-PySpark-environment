```
data=sc.textFile("/FileStore/tables/assgn2P1.txt")
```

```
data.take(10)
```

```
def parse_line(line):
    user,friends=line.split('\t')
    friends_list=friends.split(',')
    return (user, friends_list)
```

```
user_data=data.map(parse_line)
user_data.take(10)
```

```
def mutual_friend_pairs(user_friends_pair):
    user,friends=user_friends_pair
    mutual_pairs=[]
    for friend in friends:
        mutual_pairs.append(((user,friend),friends))
    return mutual_pairs
```

```
mutual_friends=user_data.flatMap(mutual_friend_pairs)
mutual_friends.take(10)
```

```
friend_counts = mutual_friends.reduceByKey(lambda list1, list2: len([friend for friend in list1 if friend in list2]))
existing_friends = user_data.flatMap(lambda x: [(x[0], f) for f in x[1]])
existing_friends.take(10)
```

```
friend_counts.take(10)
```

```
filtered_recommendations = friend_counts.leftOuterJoin(existing_friends).filter(lambda x: x[1][1] is None).map(lambda x: (x[0], x[1][0]))
filtered_recommendations.take(10)
```

```
sorted_recommendations = filtered_recommendations.map(lambda x: (x[0][0], (x[0][1], len(x[1])))).groupByKey().map(lambda x: (x[0], sorted(list(x[1]), key=lambda y: y[1], reverse=True)))
sorted_recommendations.take(10)
```

```
top_10_recommendations = sorted_recommendations.map(lambda x: (x[0], [friend for friend, count in x[1][:10]]))
top_10_recommendations.take(10)
```
